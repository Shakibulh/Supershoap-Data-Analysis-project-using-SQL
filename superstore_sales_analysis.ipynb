**1. Show the distribution of segment by pie plot.**

```python
segment_type = df['Segment'].value_counts().reset_index()
segment_type = segment_type.rename(columns={'Segment': 'Customer Type', 'count': 'Total Customers'})

segment_type
```

```python
colors = ['#ff9999','#66b3ff','#99ff99']  # light red, light blue, light green (example colors)

plt.pie(segment_type['Total Customers'],
        labels=segment_type['Customer Type'],
        autopct='%1.1f%%',
        colors=colors,
        startangle=90)  # startangle to rotate the pie chart for better look

plt.title('Distribution of categories')
plt.axis('equal')  # to make the pie chart circular
plt.show()
```

**2. Show the sales for each category.**

```python

# Assuming df is your original DataFrame

# Step 1: Prepare the data
segment_sales = df.groupby('Segment')['Sales'].sum().reset_index()
segment_sales = segment_sales.rename(columns={'Segment': 'Customer Type', 'Sales': 'Total Sales'})

# Step 2: Plotting
colors = sns.color_palette("pastel")[0:len(segment_sales)]

plt.figure(figsize=(8,5))
bars = plt.bar(segment_sales['Customer Type'], segment_sales['Total Sales'], color=colors)

plt.title('Sales Data for Each Segment', fontsize=14, fontweight='bold')
plt.xlabel('Segment Type', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{int(height):,}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

**3. Which category of shipping contains the highest percentage ratio?**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data
shipping_mode = df['Ship Mode'].value_counts().reset_index()
shipping_mode.columns = ['Mode_of_Shipment', 'Use_Frequency']

# Colors from seaborn pastel palette
colors = sns.color_palette('pastel')[0:len(shipping_mode)]

# Explode all slices slightly for better visual impact
explode = [0.05] * len(shipping_mode)

plt.figure(figsize=(8,8))

wedges, texts, autotexts = plt.pie(
    shipping_mode['Use_Frequency'],
    labels=shipping_mode['Mode_of_Shipment'],
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    explode=explode,
    pctdistance=0.85,
    shadow=True,
    textprops={'fontsize': 12, 'weight': 'bold', 'color': 'navy'}
)

# Draw circle for donut chart look
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Contribution of Shipping Modes (%)', fontsize=18, fontweight='bold')

# Add legend outside the pie chart
plt.legend(wedges, shipping_mode['Mode_of_Shipment'], title="Shipping Modes", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=12)

plt.axis('equal')  # Equal aspect ratio ensures the pie is circular.
plt.tight_layout()
plt.show()


```

**4. Show the sales for each category by bar plot?**

```python
# Group and sort as before
highest_sale_subcategory = df.groupby('Sub-Category')['Sales'].sum().reset_index()
highest_sale_subcategory = highest_sale_subcategory.sort_values(by='Sales', ascending=True)

plt.figure(figsize=(12,8))

# Color palette
colors = sns.color_palette('coolwarm', len(highest_sale_subcategory))

bars = plt.barh(
    highest_sale_subcategory['Sub-Category'],
    highest_sale_subcategory['Sales'],
    color=colors,
    edgecolor='gray',
    linewidth=0.7
)

plt.title('Sales by Sub-Category', fontsize=18, fontweight='bold')
plt.xlabel('Total Sales', fontsize=14)
plt.ylabel('Sub-Category', fontsize=14)

# Grid only on x-axis with light gray color
plt.grid(axis='x', linestyle='--', alpha=0.3)

# Add value labels on each bar
for bar in bars:
    width = bar.get_width()
    plt.text(
        width + max(highest_sale_subcategory['Sales']) * 0.01,  # small offset to right
        bar.get_y() + bar.get_height()/2,
        f'{width:,.0f}',  # formatted with commas, no decimals
        va='center',
        fontsize=12,
        color='black'
    )

plt.tight_layout()
plt.show()
```

**5. Create a line chart for sales over year.**

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
year_wise_sale = df.groupby(df['Order Date'].dt.year)['Sales'].sum()
year_wise_sale = year_wise_sale.reset_index()
year_wise_sale = year_wise_sale.rename(columns={'Order Date': 'Year', 'Sales':'Total Sales'})
year_wise_sale
plt.plot(year_wise_sale['Year'], year_wise_sale['Total Sales'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total sales by year')
plt.tight_layout()
plt.show()
```

```python
sns.set_style("whitegrid")  # nicer grid background

df["Monthly_year"] = df["Order Date"].dt.to_period("M")
monthly_sales = df.groupby("Monthly_year")['Sales'].sum()

plt.figure(figsize=(14,7))
plt.plot(
    monthly_sales.index.astype(str),
    monthly_sales.values,
    marker='o',
    color='#1f77b4',  # a nice blue color
    linewidth=3,
    markerfacecolor='white',
    markeredgewidth=2,
    markeredgecolor='#1f77b4'
)
plt.title("Sales Growth Over Time", fontsize=18, fontweight='bold')
plt.xlabel("Month-Year", fontsize=14)
plt.ylabel("Total Sales", fontsize=14)
plt.xticks(rotation=90, fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


```
